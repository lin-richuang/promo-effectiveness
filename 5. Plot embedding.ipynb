# Visualization of embedding

```python
import pandas as pd
import pickle
import matplotlib.pyplot as plt
from sklearn import manifold
import warnings

warnings.simplefilter("ignore")
pd.options.display.max_columns = None
```

```python
saved_embeddings_fname = "models/embeddings.pickle" 
label_encoder_pickle = 'models/les.pickle'
```

# Read embedding

```python
with open(saved_embeddings_fname, 'rb') as f:
    [store_embedding, year_embedding, weekofyear, dow_embedding, holiday_embedding] = pickle.load(f)
```

```python
# Load LabelEncoders
with open(label_encoder_pickle, 'rb') as f:
    les = pickle.load(f)
```

```python
len(les)
```

```python
le_store = les[0]
le_year = les[1]
le_week = les[2]
le_dow = les[3]
le_holiday = les[4]
```

# PCA 

```python
# Store
tsne = manifold.TSNE(init='pca', random_state=0, method='exact', perplexity=5, learning_rate=100)
Y = tsne.fit_transform(store_embedding)
plt.figure(figsize=(20,20))
plt.scatter(-Y[:, 0], -Y[:, 1])
for i, txt in enumerate(le_store.classes_):
    plt.annotate(txt, (-Y[i, 0],-Y[i, 1]), xytext = (-20, 8), textcoords = 'offset points')
plt.savefig('store_embedding.pdf')
```

```python
# Week of year
tsne = manifold.TSNE(init='pca', random_state=0, method='exact', perplexity=5, learning_rate=100)
Y = tsne.fit_transform(weekofyear)
plt.figure(figsize=(8,8))
plt.scatter(-Y[:, 0], -Y[:, 1])
for i, txt in enumerate(le_week.classes_):
    plt.annotate(txt, (-Y[i, 0],-Y[i, 1]), xytext = (-20, 8), textcoords = 'offset points')
plt.savefig('wek_of_year_embedding.pdf')
```

```python
# Day of week
tsne = manifold.TSNE(init='pca', random_state=0, method='exact', perplexity=5, learning_rate=100)
Y = tsne.fit_transform(dow_embedding)
plt.figure(figsize=(8,8))
plt.scatter(-Y[:, 0], -Y[:, 1])
for i, txt in enumerate(le_dow.classes_):
    plt.annotate(txt, (-Y[i, 0],-Y[i, 1]), xytext = (-20, 8), textcoords = 'offset points')
plt.savefig('day_of_week_embedding.pdf')
```

```python
# holiday
tsne = manifold.TSNE(init='pca', random_state=0, method='exact', perplexity=5, learning_rate=100)
Y = tsne.fit_transform(holiday_embedding)
plt.figure(figsize=(8,8))
plt.scatter(-Y[:, 0], -Y[:, 1])
for i, txt in enumerate(le_holiday.classes_):
    plt.annotate(txt, (-Y[i, 0],-Y[i, 1]), xytext = (-20, 8), textcoords = 'offset points')
plt.savefig('holiday_embedding.pdf')
```
